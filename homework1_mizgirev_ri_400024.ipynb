**Polars**

```python
import pandas as pd
import numpy as np
import polars as pl
```

```python
df = pl.read_csv("train.csv")
```

```python
df
```

```python
df.describe()
```

```python
passenger_count_by_class = df.groupby('Pclass').agg(pl.col('Pclass').count().alias('PassengerCount'))
```

```python
print(passenger_count_by_class)
```

```python
survived_counts = df.filter(df['Survived'] == 1).groupby('Sex').agg(pl.col('Sex').count().alias('Survived_Count'))
```

```python
print(survived_counts)
```

```python
filtered_df = df.filter(df['Age'] > 44)
```

```python
print(filtered_df)
```

**Оптимизация типов pandas**

```python
df2 = pd.read_csv("Housing.csv")
df2
```

```python
df2.info()
#Вывожу общую информацию о типах данных и использовании памяти
```

```python
df2.describe(include='all')
#Анализ статистики по числовым данным
```

```python
df2_optimized = df2.copy()
#Оптимизация типов данных
```

```python
numeric_cols = df2_optimized.select_dtypes(include=['number']).columns
df2_optimized[numeric_cols] = df2_optimized[numeric_cols].apply(pd.to_numeric, downcast='integer')
#оптимизация для числовых столбцов
#Использовал downcast='integer' при преобразовании числовых столбцов (такие как, price, area, bedrooms, bathrooms, stories, parking) для уменьшения использования памяти, сохраняя целочисленные типы данных.
```

```python
categorical_cols = df2_optimized.select_dtypes(include=['object']).columns
df2_optimized[categorical_cols] = df2_optimized[categorical_cols].astype('category')
#оптимизация для категориальных столбцов
#Преобразовал столбцы с категориальными данными (такие как, mainroad, guestroom, basement, hotwaterheating, airconditioning, prefarea, furnishingstatus) в тип данных "категория". Это позволило сократить использование памяти.
```
